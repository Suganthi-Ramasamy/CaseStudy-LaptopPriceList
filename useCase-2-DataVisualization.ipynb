# Usecase:2 Categorical Data Visualization


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
pricelist = pd.read_csv('C:/Users/Suganthi_Ramasamy/OneDrive - Dell Technologies/Desktop/DS/Data/Pricelist.csv', sep=",")
pricelist.head()
```

Bar Plot of Laptop Brands by Count

```python
df = pd.DataFrame(pricelist, columns=['brand'])
ax = df['brand'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Count of Laptops by Brand")
ax.set_xlabel("Brand")
ax.set_ylabel("Count")
plt.show()
```

```python
# Count the number of laptops for each brand
brand_counts = pricelist['brand'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
brand_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Count of Laptops by Brand', fontsize=16)
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

Pie Chart of Processor Brand Proportion
(show the distribution of laptops by processor brands)

```python
processor_brand_counts = pricelist['processor_brand'].value_counts()
processor_brand_counts.plot(kind='pie')
plt.title('Count of Laptops by Processor Brand', fontsize=16)

plt.show()
```

Countplot of Laptops with Touchscreen (Count the number of laptops with and without a touchscreen)

```python
plt.figure(figsize=(10, 6))
sns.countplot( x="brand", data=pricelist, hue="Touchscreen")
plt.title('Count of Laptops by Screen Type', fontsize=16)
#plt.show()
```
