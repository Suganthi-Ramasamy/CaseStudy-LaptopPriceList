# Correlation Heatmap
Relationships between numerical variables like Price, RAM, and SSD


First find out the numerical columns in the data list

```python
pricelist.select_dtypes(include="number").columns
```

Find the correlation between the numerical datas

```python
pricelist.select_dtypes(include="number").corr()
```

Store the correlation in a variable to get the heatmap nad draw the heatmap

```python
data = pricelist.select_dtypes(include="number").corr()
plt.figure(figsize= (10,8))
plt.title('Correlation Coefficient')
sns.heatmap(data, annot=True)
plt.show()
```
