Unique values and their counts

```python
for i in pricelist.columns:
  if pricelist[i].dtype == 'object':
    print("Unique Count in {} column: {}\nThe unique value: {}".format(i, len(pricelist[i].unique()), pricelist[i].unique()))
    print(pricelist[i].value_counts())
    print("---------------------------------------------------------------------------------\n")
```

```python
for i in pricelist.columns:
  if pricelist[i].dtype == 'object':
    print("Unique Count in column {} and its values: {}\nThe unique values: {}".format(i, len(pricelist[i].unique()), pricelist[i].unique()))
    print(pricelist[i].value_counts())
    print("--------------------------------------------\n")
  elif pricelist[i].dtype == 'int64' and len(pricelist[i].unique()) <= 20:
     print("Unique Count in column {} and its values: {}\nThe unique value: {}".format(i, len(pricelist[i].unique()), pricelist[i].unique()))
     print(pricelist[i].value_counts())
     print("--------------------------------------------\n")
```

```python
summary_stats = pricelist.describe()

for col in summary_stats.columns:
    print(f"Summary Statistics for column '{col}':")
    for stat in summary_stats.index:
        print(f"  {stat}: {summary_stats.at[stat, col]}")
    print("--------------------------------------\n")  
```
